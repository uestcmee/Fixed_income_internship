```python
import pandas as pd
# 设置绘图
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()
#设置字体为SimHei显示中文
plt.rcParams['font.sans-serif'] = 'SimHei'
#设置正常显示字符
plt.rcParams['axes.unicode_minus'] = False
```

```python
df=pd.read_excel('城投债大全.xls')[:-1]
df.head(1)
```

```python
df.index=pd.DatetimeIndex(df['起息日'])
df.head(1)
```

```python
df_small=df[df['发行方式']=='公募']
df_small
```

```python
df['2021-1']['起息日'].value_counts().sort_index()
```

```python
month2020=pd.DataFrame(columns=['month','num','amount'])
for month in range(1,13):
    month_df=df['2020-{}'.format(month)]
    month2020.loc['2020-{}'.format(month)]=([month,len(month_df),month_df['发行规模(亿元)'].sum()])
    
month_df=df['2021-1']
month2020.loc['2021-1']=([1,len(month_df),month_df['发行规模(亿元)'].sum()])
month2020
```

```python
month2020['num'].plot.bar()
```

```python
month2020['amount'].plot.bar()
```

```python
a=[30034.30,37830.45,52317.91,48732.29,47628.20,42525.23,48219.69,59805.22,56715.83,50270.87,48191.48,46684.58,26890.67]
b=[11545.09,12739.39,24781.70,24903.05,15461.86,15499.76,15681.14,19584.05,19349.81,15016.44,15741.98,11548.87,12416.63]

month2020['tot']=a
month2020['tot_credit']=b
month2020
```

```python
pct=month2020['amount']/month2020['tot']
pct.plot.bar(title='城投发行占（所有债券）发行比例')
```
