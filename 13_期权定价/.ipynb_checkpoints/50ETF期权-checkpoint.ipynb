## 50ETF基金数据获取波动情况（30日）

```python
year_trade_day=252

df_etf=pd.read_excel('50ETF_日线数据.xlsx',index_col=0).iloc[:-7]
df_etf.index=df_etf['交易时间']
df_etf['std']=df_etf['涨跌幅%'].rolling(30).std() # 日度涨跌幅的三十日标准差

std=df_etf['2020-06':'2020-08']['std']/100*np.sqrt(year_trade_day)  # 年化波动率（由于交易日数据，因此使用交易日数量来进行年化

std.plot() # 30日滚动波动率
plt.rcParams['font.sans-serif']=['SimHei']    # 用来设置字体样式以正常显示中文标签
plt.rcParams['axes.unicode_minus']=False    # 默认是使用Unicode负号，设置正常显示字符，如正常显示负号
plt.title('波动率时序图')
plt.show()
```

```python
due=datetime.datetime.strptime('2020-12-23','%Y-%m-%d')
days=(due-df.index).days
t=days/365  # 到期时间（年  # 距离到期还有多少年
r=df['Shibor1Y收盘价']/100 # 无风险利率
sigma = std  # 标准差
strike = 2.8 # 执行价格
spot = df['50ETF基金']  # 现价
```

```python
def blackScholes(s,x,t,vol,r):
    d1=(np.log(s/float(x))+(r+0.5*(vol**2))*t)/(np.sqrt(t)*(vol))
    d2=d1-vol*np.sqrt(t)
    ret=s*norm.cdf(d1)-x*(np.exp(-r*t)*norm.cdf(d2))
    return ret

opt_pred=blackScholes(spot,strike,t,sigma,r)
```

```python
df['BS预测价格']=opt_pred
opt_pred.plot(label='预测期权价格',figsize=(10,7))
df['50ETF期权'].plot()
plt.legend()
plt.title('BS定价与实际价格')
plt.show()
```

## 二叉树定价模型

```python

def BinomalTree(sigma,r,steps,spot,dt):
#     dt=1/252 # 每一步的步长（步长为一个交易日）
    r=r/100 # 除以100去掉百分号
    up = np.exp((r  - 0.5*sigma*sigma)*dt + sigma*np.sqrt(dt))
    down = np.exp((r  - 0.5*sigma*sigma)*dt - sigma*np.sqrt(dt))

    discount = np.exp(-r*dt) #dt表示为日度波动率
    steps=int(steps)
    
#     print(up**365,down**365,steps,discount**365)
    stock=np.zeros((steps,steps))
    stock[0][0]=spot
    for i in range(1,steps):
        for j in range(0,i+1):
            if j==0:
                stock[i][j]=stock[i-1][j]*down
            else:
                stock[i][j]=stock[i-1][j-1]*up

    opt_tree=(stock-strike).clip(min=0) # 每一个节点行权后的收益
    opt_price_tree=opt_tree
    for i in range(steps-2,-1,-1):
        for j in range(0,i+1):
            # 逆推得到初始点的期权价格
            opt_price_tree[i][j]=(opt_price_tree[i+1][j]+opt_price_tree[i+1][j+1])/2*discount

    return opt_price_tree

# 计算各个时间点的期权预测价格
opt_list=[]
tSteps=pd.Series(days*(252/365)).apply(lambda x:round(x,0)) # 估计有多少个交易日，即生成多少步二叉树
for i in range(len(df)):
    opt_price=BinomalTree(sigma[i],r[i],tSteps[i],spot[i],dt=1/252)
    opt_list.append(opt_price[0][0])
    
```

```python
df['二叉树定价']=opt_list
plt.figure(figsize=(10,7))
plt.plot(df[['50ETF期权','二叉树定价']])
plt.legend(['50ETF期权','二叉树定价'])
plt.rcParams['font.sans-serif']=['SimHei']    # 用来设置字体样式以正常显示中文标签
plt.rcParams['axes.unicode_minus']=False    # 默认是使用Unicode负号，设置正常显示字符，如正常显示负号
plt.title('二叉树定价与实际价格')
plt.show()
```

```python

df['二叉树定价']=opt_list
plt.figure(figsize=(10,7))
plt.plot(df[['50ETF期权','二叉树定价','BS预测价格']])
plt.legend(['50ETF期权','二叉树定价','BS预测价格'])
plt.title('两种股价方法与期权实际价格对比')
plt.show()
```

```python
df['二叉树定价']=opt_list
plt.figure(figsize=(10,7))

diff=df['二叉树定价']-df['BS预测价格']
print(diff.mean())
plt.plot(diff)

plt.title('两种估价方法价格之差')
plt.show()
```

## 两种不同步长的比较

1. 步长使用**交易日**，步数也使用交易日
2. 步长使用**日历日**，步数也使用日历日

两种方式只是划分精度不同，但是价格不应该出现较大差距，可以从下列结果中得到验证（所以之前BS&二叉树价格很近是算错了？）

```python

# 计算各个时间点的期权预测价格
opt_list2=[]
tSteps2=days # 估计有多少个交易日，即生成多少步二叉树
for i in range(len(df)):
    opt_price=BinomalTree(sigma[i],r[i],tSteps2[i],spot[i],dt=1/365)
    opt_list2.append(opt_price[0][0])
df['二叉树定价2']=opt_list2
```
